## R1

```
pet = ['cat','cook','porg']
```

```
words = [0]*26
```

```
for v in pet:
    for w in v:
        words[ord(w)-97] = words[ord(w)-97]+1
words
```

```
A = 0
Max = 0
MaxD = 0
for i in range(len(words)):
    if words[i]>0: A = A + 1
    if words[i] > Max: 
        MaxD = i
        Max = words[i]

print("[1] There are " + format(A) + " different words")
print("[2,3] Word  \"" + chr(MaxD+97) + "\" appear "+format(words[MaxD])+" times")
```
