##  E1 .1

```
def carbon_foot(typ, km):
    if typ =="car":
        return 0.24*km
    elif typ =="bus":
        return 0.03*km
    elif typ =="moto":
        return 0.06*km
    else:
        return None
```

```
print("Carbon foot : ",carbon_foot("car",100))
```

## E1.2

```
def sal(score,overwork,original):
    total = 0
    if score>=90:
        total = total + 8000
    elif score>=70 and score <90:
        total = total + 6000
    elif score <70:
        total = total + 4000
    return total + overwork*200 + original
```

```
print("A get : ",sal(90,10,30000))
print("B get : ",sal(60,5,40000))
print("C get : ",sal(80,10,50000))
```

## E1.3

```
def square(m):
    print("square high : ",m)
    for h in range(1,m+1):
        for w in range(1,m+1):
            print(h,"*",w,"=",w*h,end="\t")
        print("\n")
```

```
m = int(input("Enter how much you want : "))
square(m)
```

```
import random
```

```
def set_answer():
    ans = []
    for i in range(4):    
        ans.append(random.randint(1,9))
    return ans 
```

```
def compair(ANS,ENT):
    leng = len(ANS)
    ans = ANS.copy()
    ent = ENT.copy()
    #print(leng)
    A = 0
    B = 0
    for i in range(leng):
        if ans[i] == ent[i]:
            A = A + 1
            ans[i] = None
            ent[i] = None
    if A == 4:
        return True,A,B
    for i in range(leng):
        if ans[i]!=None:
            for u in range(leng):
                if ent[u]!=None and ent[u]== ans[i]:
                    B = B + 1
    return False,A,B
        
    
```

```
done = False
first = True
ans = []

while not done:
    if first:
        ans = set_answer()
        print(ans)
        first = False
    inp = input("Guess 4 numbers : ")
    while len(inp)!=4:
        print("Error")
        inp = input("Guess 4 numbers : ")
    #inp = int(inp)
    inpC = []
    for c in inp:
        inpC.append(int(c))
    #print(type(ans[0]))
    done,a,b = compair(ans,inpC)
    print(a,"A",b,"B")
    
print("Congratulations!!!")
    
```
